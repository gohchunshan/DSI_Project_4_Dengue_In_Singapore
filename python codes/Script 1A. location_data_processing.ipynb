```python
#import library
import pandas as pd
import os
import glob
```

```python
#create blank dataframe
df_main = pd.DataFrame(columns=['number_of_cases', 
                           'street_address', 
                           'latitude', 'longitude',
                           'cluster_number',
                           'recent_cases_in_cluster',
                           'total_cases_in_cluster',
                           'data_date',
                           'data_month_number'])
```

```python
#read multiple csv files and combined them
files = glob.glob('sgcharts//*.csv', recursive=False)
for names in files:
    print(names)
    df = pd.read_csv(names,  names=['number_of_cases', 'street_address', 'latitude', 'longitude','cluster_number','recent_cases_in_cluster','total_cases_in_cluster','data_date','data_month_number'], low_memory=False)
    print(df.shape[0])
    df_main = pd.concat([df_main, df], ignore_index=True, axis=0)
```

```python
location_df = pd.read_csv('location_replacement.csv')
my_dict = location_df.set_index('keyword').to_dict()['replace']
```

```python
def clean_address(text):
    text = text.lower()
      
    for i,j in my_dict.items():
        if text.find(i) != -1:
            return j
            break
        
    return 'none'
```

```python
df_main['area'] = df_main.street_address.apply(clean_address)
```
