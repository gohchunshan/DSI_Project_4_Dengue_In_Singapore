```python
df.dropna(subset=['Year', 'Month', 'Day'], inplace=True)
```

```python
df = df.replace(u"\x97", "")
```

```python
df = df.replace('', np.nan)
```

```python
df.isnull().sum()
```

```python
df = df.astype({"Year":'int', "Month":'int', "Day":'int'}) 
```

```python
df = df.astype({"Year":'str', "Month":'str', "Day":'str'}) 
```

```python
df["Date"] = df["Year"] + "/" + df["Month"] + "/" + df["Day"]
df
```

```python
df.to_csv('../data/weather_cleaned.csv', index=False)
```

```python
df['Date'] = pd.to_datetime(df['Date'])  
# df['Date1'] = df.apply(lambda x: x['Date'] - timedelta(days=1), axis=1)
df['daysoffset'] = df['Date'].apply(lambda x: x.weekday(6))
df['start'] = df.apply(lambda x: x['Date'] - timedelta(days=x['daysoffset']), axis=1)
df['end'] = df.apply(lambda x: x['start'] + timedelta(days=6), axis=1)
df['week'] = pd.to_datetime(df['start']).apply(lambda x: x.weekofyear())
```

```python
from datetime import datetime, timedelta

day = '01/01/2022'
dt = datetime.strptime(day, '%d/%m/%Y')
start = dt - timedelta(days=dt.weekday())
end = start + timedelta(days=6)

print("Week Start Date:"+str(start))
print("Week End Date:"+str(end))
```
